```python
import pickle
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans  
import matplotlib.pyplot as plt
```

```python
with open('dna_embedding.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
cluster_list = list()
```

```python
for i in range(10):
    tsne = TSNE(n_components=2, n_iter=100000)
    X_2d = tsne.fit_transform(data)

    kmeans = KMeans(n_clusters=2)  
    kmeans.fit(X_2d)  
    cluster_list.append(kmeans.labels_)

    plt.figure(figsize=(6, 5))
    pad = 20
    plt.xlim(min(X_2d[:,0]) - pad, max(X_2d[:,0]) + pad)
    plt.ylim(min(X_2d[:,1]) - pad, max(X_2d[:,1]) + pad)
    plt.scatter(X_2d[:,0], X_2d[:,1], c=kmeans.labels_, cmap='rainbow')
    for i in range(len(data)):
        plt.annotate(
            i+1,
            xy=(X_2d[i,0], X_2d[i,1]), xytext=(0, -20),
            textcoords='offset points', ha='right', va='bottom',
            #bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
            arrowprops=dict(arrowstyle = '-', connectionstyle='arc3,rad=0')
        )
```

```python
for i in range(len(cluster_list)):
    if cluster_list[i][0] == 1:
        for j in range(len(cluster_list[i])):
            if cluster_list[i][j] == 0:
                cluster_list[i][j] = 1
            else:
                cluster_list[i][j] = 0
```

```python
for cluster in cluster_list:
    print(cluster)
```
